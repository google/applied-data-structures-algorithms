# Trees

## Lesson Overview

Trees are a specialized type of graph with a specific type of heirarchical structure. Trees can be used to represent information such as:

- A family tree
- The reporting heirarchy in a company
- The structure of a government

The reason graphs with this structure are called trees is that trees have a *root node* (A in the diagram below) and *leaf nodes* (D, E, F, and G in the diagram below), which, if you use your imagination, looks like an upside-down tree.
