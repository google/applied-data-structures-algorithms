# Graph Properties

## Lesson Overview

> A **node** is a data structure that stores information.

> An **edge** is a data structure that represents the connection between two nodes.

> A **graph** is a data structure that connects nodes to other nodes via a network of edges.

You can think of a graph as an extension of a linked list or a doubly linked list. In a linked list, the connection between nodes is only ever implicit (via `next`, and `previous` for doubly linked lists). Graphs place no restrictions on the number of edges a node can have. Furthermore, graph edges can also carry additional information about the connection between two nodes. There are many applications of graphs in the real world:

*   Graphs can represent locations (and the roads that connect them).
*   Graphs can represent people (and their friendships / social connections).
*   Graphs can represent the Internet (websites are nodes and links to other websites serve as edges).

Nodes in graphs can even be connected to themselves, which, for instance, might be a website linking to itself via the "Home" button.



### Graph implementation

There are several ways to implement a graph in Python. Some implementations include extra data stored on edges between nodes, such as a **weight** or **cost**. If we don't require that additional data, we can make the edges implicit (rather than explicitly defining an `Edge` class) by storing a reference to a node, rather than a new edge.

```
class Node:

  def __init__(self, name):
    self.name = name
    self.connected_to = []

  def add_connection(self, node):
    if node not in self.connected_to:
      self.connected_to.append(node)
    # Nodes, in our implementation, are always connected reciprocally.
    if self not in node.connected_to:
      node.connected_to.append(self)

  def __str__(self):
    return '%s: %s' % (self.name, [node.name for node in self.connected_to])

  def __repr__(self):
    return '%s: %s' % (self.name, [node.name for node in self.connected_to])
```

```
class Graph:

  def __init__(self):
    self.node_list = []

  def add_node(self, node):
    self.node_list.append(node)

  def remove_node(self, node):
    self.node_list.remove(node)

  def print(self):
    for node in self.node_list:
      print(node)
```

Let's take a look at an example:

```
node_one = Node(1)
node_two = Node(2)
node_three = Node(3)
node_four = Node(4)
node_five = Node(5)

graph = Graph()
graph.add_node(node_one)
graph.add_node(node_two)
graph.add_node(node_three)
graph.add_node(node_four)
graph.add_node(node_five)

node_one.add_connection(node_four)
node_two.add_connection(node_four)
node_two.add_connection(node_five)
node_four.add_connection(node_five)

graph.print()
```

This code creates a graph with five nodes, some of which are connected to one or other nodes in the graph.

- Node 1 is connected to node 4
- Node 2 is connected to nodes 4 and 5
- Node 3 is not connected to any other nodes
- Node 4 is connected to nodes 1, 2, and 5
- Node 5 is connected to nodes 2 and 4

You'll notice node 3 is not connected to any other node; that's fine! It's still part of the graph. There are no restrictions on how many nodes a node should be connected to, nor on how many nodes in a graph, nor on how many edges in a graph.

### Solution

Nodedge has conflated a few things, here: they've made it so that already-sleeping students don't count towards the total number of connections! Since they subtract `num_sleeping_students` from the number of nodes in `get_students`, they're ignoring students that are no longer awake, which is not what the problem requires. They've also accidentally used `self.get_students` instead of `student.connected_to`, which just checks the size of the graph rather than the degree of each node! Use `student.connected_to` to check the degree, and then add them to the list if it's below the threshold.

```
REQUIRED_NUM_CONNECTIONS = 2


class Student(Node):
  
  def __init__(self, name, student_id):
    super(Student, self).__init__(name)
    self.id = student_id
    self.is_sleeping = False


class Homeroom(Graph):

  def add_student(self, student):
    self.add_node(student)

  def remove_student(self, student):
    self.remove_node(student)

  def get_students(self):
    return self.node_list

  def get_most_connected_students(self):
    connections_map = {}
    max_connections = -1
    for student in self.get_students():
      num_connections = student.get_degree()
      if num_connections in connections_map:
        connections_map[num_connections].append(student)
      else:
        connections_map[num_connections] = [student]
      if num_connections > max_connections:
        max_connections = num_connections
    return connections_map[max_connections]

  def get_least_connected_students(self):
    connections_map = {}
    min_connections = len(self.get_students())
    for student in self.get_students():
      num_connections = student.get_degree()
      if num_connections in connections_map:
        connections_map[num_connections].append(student)
      else:
        connections_map[num_connections] = [student]
      if num_connections < min_connections:
        min_connections = num_connections
    return connections_map[min_connections]

  def get_3_clique(self):
    # First, choose an arbitray node.
    for first_student in self.get_students():
      for second_student in first_student.connected_to:
        # When we look for students in the second student's connections, any
        # student in the first student's connections will do, but we need to
        # make sure we don't look at the first or second students!
        for third_student in second_student.connected_to:
          if (first_student in third_student.connected_to and
              first_student.id != third_student.id and 
              second_student.id != third_student.id):
            return [first_student.name, second_student.name, third_student.name]

  def check_for_sleeping_students(self):

    sleeping_students = []
    for student in self.get_students():
      if len(student.connected_to) < REQUIRED_NUM_CONNECTIONS:
        student.is_sleeping = True
        sleeping_students.append(student.name)

    return sleeping_students
```

You may wonder why students that are asleep are still counted as "connected" to other, awake students. If they were removed from the graph, that would cause the graph to change over time until it hits an equilibrium point where every student is connected to (in our case) 3 other students. That's referred to as a "3-core" (since we are checking for 3 connections), and it's a very interesting problem in graph theory with implications beyond students asleep in class. (It has a lot of interesting applications for social networks!)
