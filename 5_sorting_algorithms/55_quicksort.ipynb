# Quicksort

## Lesson Overview

Quicksort is a sorting algorithm that divides the input array into elements greater than and less than a "pivot" recursively until the entire array is sorted.

> The average case time complexity of quicksort is $O(n\log(n))$.

### Algorithm

Here is an outline of the quicksort algorithm.

1. Choose an arbitrary element in the array to be the "pivot".
1. Move all elements less than or equal to the pivot to the left of the pivot. Move all elements greater than the pivot to the right of the pivot.
1. Recursively apply this process to the sub-arrays to the left and right of the pivot.
1. Stop when a sub-array has length 0 or 1 (since it must be sorted).

### Solution

Your teammate is absolutely right that quicksort, when using the first or last element as the pivot, performs worse on arrays that are already sorted, or close to sorted. Moreover, choosing a random pivot is a clever and common workaround.

The issue with the code is a minor bug. When using an arbitrary pivot index, the first parameter passed to `pivot` should be the sub-array containing all elements in `arr` *except* `arr[pivot_index]`, whereas the code above passes the sub-array containing all elements of `arr` *up to* `arr[pivot_index]`.

To fix this, the line `left, right = pivot(arr[:pivot_index], arr[pivot_index])` should be changed to one of the following. There are a few ways to change this.

```python
# Using pop()
new_arr = arr.copy()
new_arr.pop(pivot_index)
left, right = pivot(new_arr, arr[pivot_index])
```

```python
# Using slices
left, right = pivot(arr[:pivot_index] + arr[(pivot_index + 1):], arr[pivot_index])
```

```python
# Using list comprehension
left, right = pivot([arr[i] for i in range(len(arr)) if i != pivot_index], arr[pivot_index])
```

Each has advantages and disadvantages.

- Using `pop()` is easier to understand, but has to reallocate another array `new_arr` to storage.
- Using slices may be harder to follow for the reader, but it is the simplest implementation, and works across any language.
- Using list comprehension is very Pythonic (which could be advantageous or disadvantageous depending on what you want) and is very easy to understand, but the resulting line is very long.

### Hint

The worst case occurs when `pivot` splits the array of length $n$ into sub-arrays of length of $n-1$ and 1. Let $T(n)$ be the time complexity of quicksort. Can you find a way to relate $T(n)$ to $T(n-1)$?

### Solution

Quicksort works best when `pivot` divides the sub-arrays into relatively even groups. This allows the algorithm to work recursively on smaller and smaller arrays. On the contrary, quicksort is least efficient when the pivot value is the greatest/smallest in the array, so that either `left` or `right` ends up with all of the elements.

Let $T(n)$ be the time complexity of quicksort for an array of $n$ elements. In the first iteration of quicksort, assuming the worst case, `pivot` separates the array of length $n$ into an array of length $n-1$ and an empty array. As per Guided Exercise 2, this requires $n$ iterations. Therefore

$$T(n) = T(n-1) + n.$$

Assuming the worst case at every iteration, it follows that, for all $m$ we have

$$T(m) = T(m-1) + m.$$

Therefore we have

\begin{align*}
T(n) &= T(n-1) + n \\
&= T(n-2) + (n-1) + n \\
&= T(n-3) + (n-2) + (n-1) + n \\
&\hspace{0.2cm} \vdots \\
&= T(0) + 1 + 2 + \ldots + (n-2) + (n-1) + n \\
&= \sum_{i=0}^n i \\
&= \frac{n(n+1)}{2} \\
&= \frac{1}{2} n^2 + \frac{1}{2} n \\
&= O(n^2). \\
\end{align*}

### Hint

The best case occurs when `pivot` splits the array of length $n$ into two sub-arrays of length $\frac{n}{2}$. Let $T(n)$ be the time complexity of quicksort. Can you find a way to relate $T(n)$ to $T\left( \frac{n}{2} \right)$?

### Solution

In the best case, `pivot` cuts the array of length $n$ into two arrays of length $\frac{n}{2}$ (approximately and depending on whether $n$ is odd or even). As per Guided Exercise 2, this requires $n$ iterations. Therefore

\begin{align*}
T(n) &= T \left( \frac{n}{2} \right) + T \left( \frac{n}{2} \right) + n \\
&= 2T \left( \frac{n}{2} \right) + n \\
\frac{T(n)}{n} &= \frac{2T \left(\frac{n}{2}\right)}{n} + 1 \\
&= \frac{T \left(\frac{n}{2}\right)}{\frac{n}{2}} + 1. \\
\end{align*}

Similarly,

\begin{align*}
T \left( \frac{n}{2} \right) &= \frac{T \left(\frac{n}{4}\right)}{\frac{n}{4}} + 1 \\
T \left( \frac{n}{4} \right) &= \frac{T \left(\frac{n}{8}\right)}{\frac{n}{8}} + 1. \\
\end{align*}

This continues down to $n = 2$, and

$$\frac{T(2)}{2} = \frac{T(1)}{1} + 1.$$

Since an array of length 1 is already sorted, $T(1) = 0$, so

\begin{align*}
\frac{T(2)}{2} &= \frac{0}{1} + 1 \\
&= 1. \\
\end{align*}

Continuing this idea yields

\begin{align*}
\frac{T(4)}{4} &= \frac{T(2)}{2} + 1 \\
&= 1 + 1 \\
&= 2, \\
\frac{T(8)}{8} &= \frac{T(4)}{4} + 1 \\
&= 2 + 1 \\
&= 3, \\
\end{align*}

and

\begin{align*}
\frac{T(16)}{16} &= \frac{T(8)}{8} + 1 \\
&= 3 + 1 \\
&= 4. \\
\end{align*}

The pattern here is that as $n$ increases by a factor of 2, $\frac{T(n)}{n}$ increases by 1, so 

$$2^{\frac{T(n)}{n}} = n.$$

Using log laws we have

\begin{align*}
\frac{T(n)}{n} &= \log_2(n), \text{ so} \\
T(n) &= n\log_2(n). \\
\end{align*}
