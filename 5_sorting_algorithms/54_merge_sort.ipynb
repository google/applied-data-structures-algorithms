# Merge Sort

## Lesson Overview

**Merge sort** is a sorting algorithm that works by dividing the input array into "singleton" arrays (arrays with only one element), then recursively merging them back together in such a way that order is maintained.

> The average case time complexity of merge sort is $O(n\log(n))$.

Suppose you have a function `merge` that can merge two sorted arrays into one sorted array. Merge sort works by breaking up the input array into arrays of one element (which are by definition sorted), then using `merge` to merge them back together such that the merged array is sorted.

### Algorithm

Merge sort is almost always implemented using recursion. Here is an outline of the algorithm.

1. Break the input array into two (roughly) equal arrays. (These sub-arrays do not need to be equal in length, but the algorithm is more efficient when they are roughly the same size.)
1. Sort each of them recursively using merge sort.
1. Merge the two arrays using `merge`.

### Solution

````
def merge(arr1, arr2):
  """Merges two sorted arrays into one such that the final array is sorted.
  
  Args:
    arr1, arr2: Sorted arrays.

  Returns:
    One sorted array containing all the elements from arr1 and arr2.
  """
  
  output = []
  i, j = 0, 0
  
  # Loop through arr1 and arr2 for indices that are in common.
  while i < len(arr1) and j < len(arr2):
    # Add the smaller element to the output.
    if arr1[i] < arr2[j]:
      output.append(arr1[i])
      i += 1
    else:
      output.append(arr2[j])
      j += 1
  
  # Add the rest of arr1, if any, (which is already sorted) to output.
  while i < len(arr1):
    output.append(arr1[i])
    i += 1
  
  # Add the rest of arr2, if any, (which is already sorted) to output.
  while j < len(arr2):
    output.append(arr2[j])
    j += 1
  
  return output
````

### Solution

Since all of the computations in `merge` are $O(1)$, the time complexity is the number of iterations. Since a single element of either `arr1` or `arr2` is added to `output` at each iteration, the number of iterations is `len(arr1) + len(arr2)`. This is the same in all cases. So if `arr1` has $m$ elements and `arr2` has $p$ elements, the time complexity is $O(m+p)$, in all cases.

### Solution

````
def merge_sort(arr):
  """Sorts an array of integers in ascending order."""
  if len(arr) < 2:
    return arr

  # Split the array into two.
  midpoint = len(arr) // 2
  left = arr[:midpoint]
  right = arr[midpoint:]

  # Merge the arrays recursively.
  return merge(merge_sort(left), merge_sort(right))
````

### Solution

There are two issues, one that is producing the error and one that is hidden.

Firstly, the issue that is producing the error is that `first1` and `first2` cannot be compared using `<` as these objects are instances of the `PlaylistSong` class. Classes cannot be compared directly. Instead, this line should reference `first1.length` and `first2.length`.

Secondly, the aim of this function is to ingest two linked lists sorted *descendingly* by length, and output a linked list sorted *descendingly* by length. This is the opposite to what was done in the previous exercises, so your friend should be using `>` instead of `<`.

In summary, the line `first1 < first2` should change to `first1.length > first2.length`.

[Advanced]

Show that the time complexity of `merge_sort` is $O(n\log(n))$ in all cases.

### Solution

At each iteration, `merge_sort` cuts the array of length $n$ into two arrays of length $\frac{n}{2}$ (approximately and depending on whether $n$ is odd or even). As per a previous problem, using `merge` to put these two arrays back together uses $n$ iterations. Therefore

\begin{align*}
T(n) &= T \left( \frac{n}{2} \right) + T \left( \frac{n}{2} \right) + n \\
&= 2T \left( \frac{n}{2} \right) + n.\\
\end{align*}

So we have 

\begin{align*}
\frac{T(n)}{n} &= \frac{2T \left(\frac{n}{2}\right)}{n} + 1 \\
&= \frac{T \left(\frac{n}{2}\right)}{\frac{n}{2}} + 1. \\
\end{align*}

Similarly,

\begin{align*}
T \left( \frac{n}{2} \right) &= \frac{T \left(\frac{n}{4}\right)}{\frac{n}{4}} + 1, \text{ and} \\
T \left( \frac{n}{4} \right) &= \frac{T \left(\frac{n}{8}\right)}{\frac{n}{8}} + 1. \\
\end{align*}

This continues down to $n = 2$, and

$$\frac{T(2)}{2} = \frac{T(1)}{1} + 1.$$

Since an array of length 1 is already sorted, $T(1) = 0$, so

\begin{align*}
\frac{T(2)}{2} &= \frac{0}{1} + 1 \\
&= 1. \\
\end{align*}

Continuing this idea yields

\begin{align*}
\frac{T(4)}{4} &= \frac{T(2)}{2} + 1 \\
&= 1 + 1 \\
&= 2, \\
\frac{T(8)}{8} &= \frac{T(4)}{4} + 1 \\
&= 2 + 1 \\
&= 3, \text{ and} \\
\frac{T(16)}{16} &= \frac{T(8)}{8} + 1 \\
&= 3 + 1 \\
&= 4. \\
\end{align*}

The pattern here is that as $n$ increases by a factor of 2, $\frac{T(n)}{n}$ increases by 1, so 

$$2^{\frac{T(n)}{n}} = n.$$

Using log laws we have

\begin{align*}
\frac{T(n)}{n} &= \log_2(n), \text{ so} \\
T(n) &= n\log_2(n). \\
\end{align*}

Merge sort has the same complexity for the best, average, and worst case. This is because merge sort is guaranteed to cut the array into two roughly equal sub-arrays at each iteration, whereas the size of the two sub-arrays in quicksort depends on the value of the pivot. Merge sort is $O(n\log(n))$ for all inputs.
