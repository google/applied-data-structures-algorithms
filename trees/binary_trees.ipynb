# Binary Trees

## Lesson Overview

> A **binary tree** is a tree in which each node must have no more than two children.

Limiting a tree to having no more than two children simplifies tasks that are more complex for trees with no limit on the number of children.

> The total number of nodes in a binary tree is $2^h - 1$, where $h$ is the height of the tree. A **complete** tree has exactly $2^h - 1$ nodes.

> The number of nodes at any level of a binary tree is at most $2^n$, where $n$ is the 0-indexed level of the node.

### Binary tree example

The $0^{\textrm{th}}$ level has 1 node ($2^0$) and this tree has a total of 7 nodes ($2^3 - 1$). The tree could have fewer than 7 nodes, in which case the tree would be **incomplete**.

### Binary trees in code

Representing a binary tree requires both a `Node` class and a `BinaryTree` class.

````
class Node:

  def __init__(self, name):
    self.name = name
    self.parent = None
    # "left" and "right" here refer to the left and right children of the node.
    self.left = None
    self.right = None

  def get_parent(self):
    return self.parent

  def __str__(self):
    if not self.left and not self.right:
      return '%s' % (self.name)
    if not self.left:
      return '%s: %s' % (self.name, ['', self.right.name])
    if not self.right:
      return '%s: %s' % (self.name, [self.left.name, ''])
    return '%s: %s' % (self.name, [self.left.name, self.right.name])

  def __repr__(self):
    if not self.left and not self.right:
      return '%s' % (self.name)
    if not self.left:
      return '%s: %s' % (self.name, ['', self.right.name])
    if not self.right:
      return '%s: %s' % (self.name, [self.left.name, ''])
    return '%s: %s' % (self.name, [self.left.name, self.right.name])
````

## Question 1

Which of the following statements about a binary tree are true?

**a)** In a binary tree, each node must have exactly two children.

**b)** A binary tree may have up to two root nodes.

**c)** Binary trees support pre-order, in-order, and post-order traversals, since they are just a subset of general trees.

**d)** A complete binary tree always has an even number of nodes, since all nodes will have two children.

### Solution

The correct answer is **c)**.

**a)** The nodes in a binary tree have *at most* two children, not *exactly* two children.

**b)** A binary tree, like all trees, must only have one root.

**d)** A complete binary tree has $2^h - 1$ nodes, where h is the height of the tree. This number will always be either odd or 0. Another way to think about it is that every node in a complete binary tree has two children, but the tree has one root node, so the node count will always be odd.

````
class BinaryTree:

  def __init__(self, root_node):
    self.root = root_node

  def get_nodes_inorder(self):
    node_list = []
    self._inorder_helper(self.root, node_list)
    return node_list

  def _inorder_helper(self, current_node, node_list):
    if not current_node:
      return
    self._inorder_helper(current_node.left, node_list)
    node_list.append(current_node)
    self._inorder_helper(current_node.right, node_list)

  def size(self):
    return self._size_helper(self.root)

  def _size_helper(self, current_node):
    if not current_node:
      return 0
    left_count = self._size_helper(current_node.left)
    right_count = self._size_helper(current_node.right)
    return left_count + right_count + 1
````

````
class BinaryTree:

  def __init__(self, root_node):
    self.root = root_node

  def get_nodes_inorder(self):
    node_list = []
    self._inorder_helper(self.root, node_list)
    return node_list

  def _inorder_helper(self, current_node, node_list):
    if not current_node:
      return
    self._inorder_helper(current_node.left, node_list)
    node_list.append(current_node)
    self._inorder_helper(current_node.right, node_list)

  def size(self):
    return self._size_helper(self.root)

  def _size_helper(self, current_node):
    if not current_node:
      return 0
    left_count = self._size_helper(current_node.left)
    right_count = self._size_helper(current_node.right)
    return left_count + right_count + 1

  def to_array(self):
    node_queue = [self.root]
    result = []
    while node_queue:
      current_node = node_queue.pop(0)
      if current_node.left:
        node_queue.append(current_node.left)
      if current_node.right:
        node_queue.append(current_node.right)
      result.append(current_node.name)
    return result
````
